```python
import requests
from bs4 import BeautifulSoup
feed = "https://www.motherjones.com/topics/donald-trump/feed/"
response=requests.get(feed)
webpage=response.content
soup=BeautifulSoup(webpage,features="xml")
items=soup.find_all("item")
articles=[]
for item in items:
    link=item.find("link").text
    articles.append(link)
print(articles)
```

```python
from textblob import TextBlob
```

```python
from newspaper import Article
import nltk
import pandas as pd
from textblob import TextBlob

# Download required NLTK data
nltk.download('punkt')
nltk.download('punkt_tab')

# Initialize an empty list to store article data
data_list = []

# Iterate over the list of article URLs
for url in articles:
    try:
        # Initialize and process the article
        article = Article(url)
        article.download()
        article.parse()
        article.nlp()

        # Extract relevant information
        title = article.title
        summary = article.summary
        keywords = article.keywords
        text = article.text
        date = article.publish_date

        # Perform sentiment analysis
        text_blob = TextBlob(text)
        polarity = text_blob.polarity
        subjectivity = text_blob.subjectivity

        # Store the data as a dictionary
        data = {
            'Title': title,
            'Summary': summary,
            'Keywords': keywords,
            'Text': text,
            'Publish_Date': date,
            'Polarity': polarity,
            'Subjectivity': subjectivity
        }

        # Append the dictionary to the list
        data_list.append(data)

        # Print article details (optional, for debugging)
        print("***************************************")
        print(f"Title: {title}")
        print(f"Summary: {summary}")
        print(f"Keywords: {keywords}")
        print(f"Polarity: {polarity}")
        print(f"Subjectivity: {subjectivity}")
        print(f"Text: {text}")
        print(f"Date: {date}")
        print("***************************************")

    except Exception as e:
        print(f"Error processing {url}: {e}")

# Create a pandas DataFrame from the list of dictionaries
df = pd.DataFrame(data_list)

# Display the DataFrame
print(df)

# Optionally, save the DataFrame to a CSV file
df.to_csv('articles_data.csv', index=False)
```
